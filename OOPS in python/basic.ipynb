```python
class car :
    wheel = 4 # class variable or static variable
    def __init__(self): # constructor the instance variable define in constructor
        self.mil = 10 # instance variable
        self.com = 'BMW' # instance variable
c1 = car()  
c2 = car()


c1.mil = 8 # instance variable that can be change
c1.com = 'Audi'
c1.wheel = 5 # instance variable that can be change


print(c1.com,c1.mil,c1.wheel)
print(c2.com,c2.mil, c2.wheel)

print(id(c1))
print(id(c2))
```

```python
# Abstraction 
class test :
    def __init__(self,a,b,c,d):
        self.__a =a
        self.b=b
        self.c=c
        self.d=d
    def test_custome(self,v):
        return v-self.__a
    def __str__(self):
        return "this is my test code for abstraction"
    
```

```python
o=test(4,5,6,7)
o.test_custome(7)
```
