##### Single Inheritance

* The single inheritance is the child class inherit from parent class

```python
# parent class :-
class Alpha:
    def fun(self):
        print("Alpha class fun()")
        
# child class :- 
class Beta(Alpha):
    pass
b = Beta()
b.fun()
```

#### Multilevel Inheritance
* The multi level inheritance is the grand child inherit from child class inherit from parent class

```python
# parent class :-
class Alpha:
    def fun1(self):
        print('Inside alpha fun()')
# child class :=
class Beta(Alpha):
    def fun2(self):
        print('Inside Beta fun2()')
# grand child class :-
class Gamma(Beta):
    pass

g = Gamma()
g.fun1()
g.fun2()
print(dir(Gamma))
```

#### Multiple Inheritance
* Multiple inheritance is a have a one child class is inherit from two parents classes

```python
class Alpha:
    def fun1(self):
        print('Alpha class fun1()')

class Beta:
    def fun2(self):
        print('Beta class fun2()')
        
class Gamma(Alpha,Beta):
    pass

g = Gamma()
g.fun1()
g.fun2()
```

```python
class A:
    def fun(self):
        print('A')
class B:
    def fun(self):
        print('B')
class C:
    def fun(self):
        print('C')
class D:
    def fun(self):
        print('D')
class E(A,B):
    def fun(self):
        print('E')
class F(C,D):
    def fun(self):
        print('F')
class G(E,F):
    def fun(self):
        print('G')
    
```

```python
class customer:
    
    def __init__(self,name,ph,email):
        self.name = name
        self.ph =ph
        self.email=email
        
class platiniumcustomer(customer):
    
    
    def __init__(self, name, ph, email,plat_id):
        super().__init__(name,ph,email) # calling the parent method using the super() 
        self.plat_id = plat_id
    
    def display(self):
        print(self.__dict__)
        
def main():
    p = platiniumcustomer('srikanth',6304195295,'srikanthreddya345@gmail.com',10)
    p.display()
        
if __name__=='__main__':
    main()
```
