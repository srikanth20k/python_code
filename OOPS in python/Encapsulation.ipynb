### Encapsulation





```python
# it's a direct access to change the variable

class account():
    def __init__(self,):
        self.bal = 10000 # we can able to access the variable and change the variable
ah = account()
print(ah.bal)
ah.bal = -20000
print(ah.bal)
```





#### we have to provide the controlled access insteated of the direct access
* it can done by the getter and setter

```python
class account():
    def __init__(self):
        self.bal = 10000
    def get_bal(self): # we are providing the valid contrled access
        return self.bal
    def set_bal(self,amt): # we are providing valid access to change the variable 
        if amt >0:
            self.bal = amt
        else :
            print('Invalid Amount')
            
            
ah = account()
# print(ah.get_bal())
# ah.set_bal(-20000)
# print(ah.get_bal())


# but we can access the bal variable by using the ah.bal
ah.bal = -20000
print(ah.bal)
```

```python
class account():
    def __init__(self):
        self._bal = 10000
    def get_bal(self): # we are providing the valid contrled access _variable anme
        return self._bal
    def set_bal(self,amt): # we are providing valid access to change the variable 
        if amt >0:
            self._bal = amt
        else :
            print('Invalid Amount')
            
            
ah = account()
# print(ah.get_bal())
# ah.set_bal(-20000)
# print(ah.get_bal())


# but we can access the bal variable by using the ah._bal
ah._bal = -10000
print(ah._bal)
print(ah.__dict__)
```
